```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('store_data.csv',header = None)
store_data.head()
```

```python
store_data.shape
```

```python
records =[]
for i in range(0,300):
    records.append([str(store_data.values[i,j]) for j in range(0, 20) ]) 
    
print(records)
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence = 0.8)
association_results = list(association_rules)
association_results
```

```python
for i in association_results:
    items = str(i[0])
    if items.find('nan') ==-1:
        print(items,"support",i[1], "Confidence: ", i[2][0][2])
```
