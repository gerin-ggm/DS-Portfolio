```python
# Create a model to predict the price on the basis of Area, Bedroom and Age
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('property-price.csv')
df
```

```python
sns.regplot(x ='Area (sqr ft)',y='Price', data=df, ci=0) # ci ->confidence interval
```

```python
sns.regplot(x ='Bedroom',y='Price', data=df, ci=0)
```

```python
sns.regplot(x ='Age',y='Price', data=df, ci=0)
```

```python
X = df.drop('Price', axis = 1)
y = df.Price
```

```python
X
```

```python
y
```

```python
# Create the Regreesion Model
model= LinearRegression()
model.fit(X, y)
```

```python
# y=m1x1+c x1-> Area
# y=m2x2+c x2-> Bedroom
# y=m3x3+c x3-> Age
```

```python
model.coef_
```

```python
# m1=1146.72719364
# m2=-407035.5121687
# m3=-81083.45245812
```

```python
model.intercept_
```

```python
# Area (sqr ft) - 407035.51 * Bedroom - 81083.45 * Age + 5394206.731
```

```python
# Internal model calculation
# m1*x1*m2*x2*m3*x3+c
price = 1146.72 * 3000- 407035.51 * 4- 81083.45 * 15 + 5394206.731
price
```

```python
pred = model.predict(X)
pred
```

```python
from sklearn.metrics import r2_score
r2_score(y, pred)
```
